Voy a crear un indice de tipo fecha para poder indexar este dataframe.
Para eso vamos a crear una función, que le pasamos como parametros (año, trimestre)
y devolvemos una fecha perteneciente a ese trimestre.

```python
def obtener_fecha_trimestre(año, trimestre):
    if trimestre == 1:
        return f'{año}-01-01'  # 1 de enero
    elif trimestre == 2:
        return f'{año}-04-01'  # 1 de abril
    elif trimestre == 3:
        return f'{año}-07-01'  # 1 de julio
    elif trimestre == 4:
        return f'{año}-10-01'  # 1 de octubre
    else:
        return None  # En caso de un trimestre inválido

# Crear una nueva columna 'Fecha' aplicando la función
hoja2_agrupado['Fecha'] = hoja2_agrupado.apply(lambda row: obtener_fecha_trimestre(row['Año'], row['Trimestre']), axis=1)
# Convertir la columna 'Fecha' a tipo datetime
hoja2_agrupado['Fecha'] = pd.to_datetime(hoja2_agrupado['Fecha'])
```

```python
# Asignar la nueva columna 'fecha' como índice
hoja2_agrupado.set_index('Fecha', inplace=True)
# Mostrar el DataFrame resultante
hoja2_agrupado.columns
```

Vamos a extraer y ver los datos de la CUARTA HOJA - Velocidad % por prov
Esta se va a analizar completamente porque se observo que no contiene datos nulos.

Ademas se va acrear un indice del tipo fecha, para poder indexar con respecto a un año y a 
un trimestre.

```python
hoja4 = pd.read_excel(path_arch, sheet_name="Velocidad % por prov", engine='openpyxl')
print(hoja4.info())
hoja4.columns
hoja4.head()
```

```python
# Crear una función para generar un índice de tipo fecha 
def crear_fecha_unica(año, trimestre, dia_incremento):
    meses = {1: 1, 2: 4, 3: 7, 4: 10}
    mes = meses[trimestre]
    # Controlo que no se me pase para que no me de erroes
    if mes in [1, 3, 5, 7, 8, 10, 12]:
        max_dias = 31
    elif mes in [4, 6, 9, 11]:
        max_dias = 30
    else:  
        if (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0):
            max_dias = 29  # Por FEbrero años bisiestos
        else:
            max_dias = 28
  
    dia = min(dia_incremento, max_dias)
    return pd.Timestamp(year=año, month=mes, day=dia)

hoja4['Fecha'] = [crear_fecha_unica(row['Año'], row['Trimestre'], idx + 1) 
                   for idx, row in hoja4.iterrows()]
# Establecer la nueva columna como indice
hoja4.set_index('Fecha', inplace=True)
```

```python
hoja4.to_csv("Hoja4_Veloc%_por_prov.csv", index=True)
hoja4.to_excel("Hoja4_Veloc%_por_prov.xlsx", index=True)
```
